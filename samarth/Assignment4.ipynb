```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df,x='purpose',hue='not.fully.paid')
plt.xticks(rotation=45);
```

```python
pre_df = pd.get_dummies(df,columns=['purpose'],drop_first=True)
pre_df.head()
```

```python
from sklearn.model_selection import train_test_split

X = pre_df.drop('not.fully.paid', axis=1)
y = pre_df['not.fully.paid']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=125
)
```

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()

model.fit(X_train, y_train);
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay

y_pred = model.predict(X_test)



print(classification_report(y_test,y_pred))
```

```python
labels = ["Fully Paid", "Not fully Paid"]
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
disp.plot();
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Parameters for the normal distribution
mu = 0  # Mean
sigma = 1  # Standard deviation

# Parameters for the Gaussian distribution
mean = 0  # Mean
std_dev = 1  # Standard deviation

# Generate data for the Gaussian distribution
x_gaussian = np.linspace(-5, 5, 1000)
y_gaussian = 1 / (std_dev * np.sqrt(2 * np.pi)) * np.exp(-(x_gaussian - mean)**2 / (2 * std_dev**2))

# Plot the Gaussian distribution curve
plt.figure(figsize=(8, 6))
plt.plot(x_gaussian, y_gaussian, label='Gaussian Distribution')
plt.xlabel('X-axis')
plt.ylabel('Density')
plt.title('Gaussian Distribution')
plt.legend()
plt.grid(True)

# Show the plots
plt.show()
```
